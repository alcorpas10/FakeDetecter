```python
bulos = {}
```

```python
cont = 0
elems = driver.find_elements(By.CLASS_NAME, 'quote_link.js-card')
for i in range(2, len(elems), 2):
        title = elems[i].text
        link = elems[i].get_attribute('href')
        bulos[cont] = {"titulo" : title, "link" : link}
        cont += 1
```
